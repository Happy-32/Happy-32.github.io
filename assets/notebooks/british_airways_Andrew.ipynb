```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways/"
pages=395
page_size=50

with open('Reviews.csv', mode='w', newline='', encoding='utf-8') as file:
    writer=csv.writer(file)
    writer.writerow(['Review Date','Reviewer Name', 'Review'])

    errors=0
    # for page in tqdm(range(1, 396), desc="Scraping pages"):
    for i in range(1,pages+1):
        try:
            url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"
            response=requests.get(url).text
            soup=BeautifulSoup(response, 'html.parser')

            reviews=soup.find_all('div',class_='body')

            for review in reviews:
                    name_tag = review.find('h3')
                    name = name_tag.find('span').text.strip() if name_tag else "N/A"
                    date = name_tag.find('time').text.strip() if name_tag else "N/A"
                    comment_div = review.find('div', class_='text_content')
                    comment = comment_div.text.strip() if comment_div else "N/A"
                    writer.writerow([date, name, comment])
               
            print(f"Sucessfull Exported page {i}")


        except Exception as e:
            print(f'Page {i} processing failed: {e}')
            errors+=1

    print(f"Data Sucessfully Exported with {errors} errors")
```
